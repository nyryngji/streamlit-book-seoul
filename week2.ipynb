```python
import xmltodict
contents = xmltodict.parse(response.text)
contents
```

```python
import numpy as np
result['월'] = np.where(result['월'].astype('int')<10,'0'+result['월'],result['월'])
result['일'] = np.where(result['일'].astype('int')<10,'0'+result['일'],result['일'])
```
