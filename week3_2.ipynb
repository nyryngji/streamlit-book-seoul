```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 서울시 열린 데이터 광장
apikey = '4c736b5971616d7937317a416f4a56'

def main():
    df = pd.DataFrame()

    for i in range(20):
        url = f'http://openapi.seoul.go.kr:8088/{apikey}/json/tbLnOpendataRtmsV/{i*1000+1}/{(i+1)*1000}/'
        response = requests.get(url)
        content = response.json()
        con = content['tbLnOpendataRtmsV']['row']
        res = pd.DataFrame(con)
        df = pd.concat([df, res], axis=0)
    
    data = df.reset_index(drop=True)
    data['DEAL_YMD'] = pd.to_datetime(data['DEAL_YMD'],format=("%Y%m%d"))
    data.to_csv('seoul_real_estate.csv',index=False)

if __name__ == "__main__":
    main()
```
